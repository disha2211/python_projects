```python
import random

x=random.randint(1,100)
attempts=0
while True:
    guess=int(input("Guess the number (between 1 and 100) : "))
    attempts += 1
    if guess < x:
        print(f"Too Low! Try Again")
    elif guess > x:
        print(f"Too High! Try Again")
    elif guess == x:
        print(f"Congratulations! You guessed the number in {attempts} attempts")
        break
```
