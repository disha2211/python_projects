```python
import random

x=int(input("set MIN : "))
y=int(input("set MAX : "))
target=random.randint(x,y)
attempts=0
while True:
    guess=int(input(f"Guess the number (between {x} and {y} ) : "))
    attempts += 1
    if guess < target:
        print(f"Too Low! Try Again")
    elif guess > target:
        print(f"Too High! Try Again")
    elif guess == target:
        print(f"Congratulations! You guessed the number in {attempts} attempts")
        break
```
