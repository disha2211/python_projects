```python
import random

x=random.randint(1,100)

while True:
    guess=int(input("Guess the number (between 1 and 100) : "))
    if guess < x:
        print(f"Too Low! Try Again")
    elif guess > x:
        print(f"Too High! Try Again")
    elif guess == x:
        print(f"Congratulations! You guessed the number")
        break
```
